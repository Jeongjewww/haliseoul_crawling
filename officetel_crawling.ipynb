```python
## 필요정보

final_complex_detail_df_list = []
for complex_id in [127168, 18433, 140630, 104169, 103453, 101538, 22689, 142050, 101863, 21908]:
    url = "https://new.land.naver.com/api/complexes/{}?sameAddressGroup=false".format(complex_id)
    res = requests.get("https://new.land.naver.com/api/complexes/{}?sameAddressGroup=false".format(complex_id), headers=headers)

    data_dict = res.json()
    
    complex_detail_df = pd.Series(data_dict['complexDetail']).to_frame().T
    complex_detail_df = complex_detail_df.loc[:, ['complexNo', 'complexName', 'realEstateTypeCode', 'realEstateTypeName', 'detailAddress']]
    complex_detail_list_df = pd.DataFrame(data_dict['complexPyeongDetailList'])
    
    complex_detail_list_df = complex_detail_list_df.rename(columns={"realEstateTypeCode": "realEstateTypeCode2"})
    
    complex_detail_df = pd.concat([complex_detail_df] * len(complex_detail_list_df))
    complex_detail_df = complex_detail_df.reset_index()
        
    complex_detail_df = pd.concat([complex_detail_df, complex_detail_list_df], axis=1)
    final_complex_detail_df_list.append(complex_detail_df)

final_complex_detail_df = pd.concat(final_complex_detail_df_list)
final_complex_detail_df.drop(['index', 'supplyPyeong', 'grandPlanList', 'articleStatistics'], axis=1, inplace=True)
```

```python
# 매물 시세 받아오기

series_list = []
for _, row in final_complex_detail_df.iterrows():
    complex_num = row['complexNo']
    area_num = row['pyeongNo']
    
    url = "https://new.land.naver.com/api/complexes/{}/prices?complexNo={}&tradeType=A1&year=5&priceChartChange=false&type=table&areaNo={}".format(
        complex_num, complex_num, area_num
    )
    res = requests.get(url, headers=headers)
    data_dict =  res.json()

    try:
        series = pd.DataFrame(data_dict['marketPrices']).iloc[0]
        series['complexNo'] = complex_num
        series['pyeongNo'] = area_num
        series_list.append(series)
    except:
        continue
```

```python
price_df = pd.concat(series_list, axis=1).T
price_df.head(10)
```

```python
price_df = price_df.loc[:, ['baseYearMonthDay', 'complexNo', 'pyeongNo', 'deposit', 'rentUpperPrice']]
```

```python
merged_df = pd.merge(
    final_complex_detail_df,
    price_df,
    on=["complexNo", "pyeongNo"],
    how="left"
)
merged_df.head()
```

```python
merged_df = merged_df.iloc[:, [1, 3, 4, 6, 8, 18, 19, 20]]
merged_df.columns = ['단지명', '거주형태', '상세주소', '면적', '평명', '기준날짜', '보증금', '월세']
merged_df = merged_df.fillna('0')

### 경도, 위도값 포함해서 저장 ###
officetel_list = [
    '용산다올노블리움', '디아뜨센트럴', 
    '채원', 'KCC웰츠타워', 
    '용산큐브', '갈월에이트리움-고려', 
    '그린빌파크', '용산큐브스테이트',
    '원효', '용산파크자이'
]

latlon = [
    [37.54200526827014, 126.97038671116634], [37.54111154264995, 126.97042045534563],
    [37.53932686192583, 126.9698991995249], [37.53790365852663, 126.96893711301728],
    [37.537435667320544, 126.96822411301726], [37.541746855022446, 126.97240072650979],
    [37.54569435033189, 126.97385808233068], [37.547317219979384, 126.96996049767421],
    [37.537938688256276, 126.96523052465864], [37.536827322155126, 126.9727933400022]
]

lat=[]; lon=[]

for value in merged_df['단지명']:
    if value == officetel_list[0]:
        lat.append(latlon[0][0])
        lon.append(latlon[0][1])
    elif value == officetel_list[1]:
        lat.append(latlon[1][0])
        lon.append(latlon[1][1])
    elif value == officetel_list[2]:
        lat.append(latlon[2][0])
        lon.append(latlon[2][1])
    elif value == officetel_list[3]:
        lat.append(latlon[3][0])
        lon.append(latlon[3][1])
    elif value == officetel_list[4]:
        lat.append(latlon[4][0])
        lon.append(latlon[4][1])
    elif value == officetel_list[5]:
        lat.append(latlon[5][0])
        lon.append(latlon[5][1])
    elif value == officetel_list[6]:
        lat.append(latlon[6][0])
        lon.append(latlon[6][1])
    elif value == officetel_list[7]:
        lat.append(latlon[7][0])
        lon.append(latlon[7][1])
    elif value == officetel_list[8]:
        lat.append(latlon[8][0])
        lon.append(latlon[8][1])
    else:
        lat.append(latlon[9][0])
        lon.append(latlon[9][1])
        
merged_df['위도'] = lat
merged_df['경도'] = lon
merged_df
```

```python
#### 지도 시각화 ####
import folium
from folium.plugins import MarkerCluster

_officetel = pd.read_csv('officetel.csv', encoding='utf8')
map = folium.Map(location=[37.54200526827014, 126.97038671116634], zoom_start=20)
marker_cluster = MarkerCluster().add_to(map)

for index, col in _officetel.iterrows():
    popup_info = '<pre>단지명: '+str(col['단지명'])+'<br>거주형태: '+str(col['거주형태'])+'<br>상세주소: '+str(col['상세주소'])+'<br>면적: '+str(col['면적'])+'<br>평명: '+str(col['평명'])+'<br>기준날짜: '+str(col['기준날짜'])+'<br>보증금: '+str(col['보증금'])+'<br>월세: '+str(col['월세'])

    folium.Marker(location = [col['위도'], col['경도']],
        popup = popup_info,
        icon = folium.Icon('blue', icon='home')
    ).add_to(marker_cluster)

map
map.save('officetel.html')
```
